### Task 3

```python
import user_agents
```

```python
vis_data=pd.read_csv('visitor-interests.csv')
vis_data.head()
```

```python
data = vis_data[:50]
data.head()
```

```python
def get_info(x):
    us_agent = user_agents.parse(x)
    return pd.Series((us_agent.browser.family , us_agent.os.family , us_agent.device.family))
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
data[['Browser','OS','Device']]=data['UserAgent'].apply(get_info)
```

```python
data.head()
```

```python
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
```

```python
def extract_latlong_from_ip(ip):
    try:
        res = ip2geo.get(ip, api_key='free')
        return pd.Series((res.latitude,res.longitude))
    except:
        return None
```

```python
data[[ 'latitude', 'longitude']]=data['IP'].apply(extract_latlong_from_ip)
```
