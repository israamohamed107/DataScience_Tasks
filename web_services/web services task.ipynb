```python
import requests ,json 
import pandas as pd
```

```python
response=requests.get('https://gorest.co.in/public-api/users?_format=json&access-token=H4lYmIGLXZkDlpy_w0ZVS5rgcpDXbN7fhRRy')
data=json.loads(response.text)
data
```

```python
data=data['data']
data
```

### Make database to save the data and run some SQL on it

```python
import sqlite3 
conn=sqlite3.connect('data.db')
```

```python
# inserting data into data.bd
for dic in data:
    query='INSERT INTO data (id,name,email,gender,status,created_at,updated_at) VALUES (?,?,?,?,?,?,?)'
    data=[dic['id'],dic['name'],dic['email'],dic['gender'],dic['status'],dic['created_at'],dic['updated_at']]
    result=conn.execute(query,data)
    conn.commit()
conn.close()
```

```python
# some sql on the new DB 
conn=sqlite3.connect('data.db')
query='SELECT status ,COUNT(id) AS users_count FROM data GROUP BY status'
data=[]
res=conn.execute(query,data).fetchall()
res
```
