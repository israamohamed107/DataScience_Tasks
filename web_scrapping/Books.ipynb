```python
import requests
from bs4 import BeautifulSoup
```

```python
import sqlite3
```

```python
link='https://books.toscrape.com//'
response=requests.get(link)
response=response.text
```

```python
soup=BeautifulSoup(response,'html.parser')
```

```python
soup.find('ul',attrs={'class':'nav-list'})
```

```python
categories=soup.find('ul',attrs={'class':'nav-list'}).get_text().strip().split('\n                            \n                        \n\n\n\n                            \n                             ')
categories.remove('Books')
categories
```

```python
categories=[ ele.strip() for ele in categories]
```

```python
categories
```

```python
cat_links=[]
for ele in soup.find('ul',attrs={'class':'nav-list'}).find_all('a'):
    cat_links.append(ele.attrs.get('href'))
```

```python
cat_links=cat_links[1:]
cat_links
```

```python
len(categories)
```

```python
len(cat_links)
```

```python
con=sqlite3.connect('Books.db')
quary="INSERT INTO books (Title,Price,Rate,Description,Category) VALUES (?,?,?,?,?) "

for n in range(len(categories)):
    res=requests.get(link+cat_links[n])
    res=res.text
    soup=BeautifulSoup(res,'html.parser')
    category=categories[n]
    for i in range(len(soup.find('ol',attrs={'class':'row'}).find_all('li'))):
        rate=soup.find('ol',attrs={'class':'row'}).find_all('li')[i].find('p',attrs={'class':'star-rating'}).attrs.get('class')[-1]
        book_link=soup.find('ol',attrs={'class':'row'}).find_all('li')[i].find('h3').find('a').attrs.get('href').strip('../')#for get descrition
        title=soup.find('ol',attrs={'class':'row'}).find_all('li')[i].find('h3').find('a').attrs.get('title')
        price=float(soup.find('ol',attrs={'class':'row'}).find_all('li')[i].find('p',attrs={'class':'price_color'}).get_text()[2:]) 
        res_2=requests.get(link+'catalogue/'+book_link)
        soup_2=BeautifulSoup(res_2.text,'html.parser')
        description=soup_2.find_all('p')[3].get_text()
        data=[ ele for ele in [title,price,rate,description,category]]
        con.execute(quary,data)
        con.commit()
        
con.close()
```
