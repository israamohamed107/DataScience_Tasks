```python
import seaborn as sns
import pandas as pd
```

```python
sns.set(rc={'figure.figsize': [8, 8]}, font_scale=1.2)
```

```python
data=pd.read_csv('50ulke.csv')
data.head()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
data.head()
```

```python
data.info()
```

```python
sns.pairplot(data,height=3,aspect=1.3)
```

```python
sns.distplot(data['National Income'],color='g')
```

> most of natinal incomes are between 500:1000

```python
sns.kdeplot(data['National Income'],shade=True,color='g')
```

```python
sns.distplot(data['Growth'],color='k',bins=20)
```

```python
sns.distplot(data['Per person'],color='r',bins=15)
```

```python
sns.boxplot(x='Continent',y='National Income',data=data,palette='PRGn')
```

> largest incomes are in North America , Europe comes after it

> median of national incomes for North America about 2000

>  most of national incomes in Asia are between 500:1000

>  most of national incomes in Europe are between 500:2000

```python
sns.violinplot(x='Continent',y='National Income',data=data,palette='PRGn')
```

```python
sns.swarmplot(x='Continent',y='National Income',data=data)
```

```python
sns.boxplot(x='Continent',y='Per person',data=data,palette='Set1')
```

> most of North Amireca incomes per person between 3000:5600

> most of North Asia incomes per person between 400:5500

> most of North Europe incomes per person between 1800:5600

```python
sns.violinplot(x='Continent',y='Per person',data=data,palette='Set1')
```

```python
sns.swarmplot(x='Continent',y='Per person',data=data,palette='Set1')
```

```python
sns.boxplot(x='Continent',y='Growth',data=data,palette='rocket')
```

> Asia has the largest growth ratio , most of Growth ratios between 3.2 : 6.8

> Europe comes after Asia , most of Growth ratios between 1.8 : 3

> Africa comes After them , most of Growth ratios between 1 : 2.8
