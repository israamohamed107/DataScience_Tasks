### Write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.


```python
conn=sqlite3.connect('books.db')
```

```python
query='SELECT books.title,categories.name ,books.rate ,books.price FROM books INNER JOIN categories ON books.category_id = categories.id'
data=[]
res=conn.execute(query,data).fetchall()
res
```

```python
for  book_name , category_name , book_rate , book_price in res :
    print(f'*) Book name is {book_name} From Category "{category_name}" with price = {book_price} has rate {book_rate}')
```

```python
conn.close()
```

### Write a sql query to calculate how many books each rate has and have price between 20£ and 30£.


```python
conn=sqlite3.connect('books.db')
```

```python
query='SELECT rate,COUNT(id) AS numer_of_books FROM books  WHERE price BETWEEN ? AND ? GROUP BY rate'
data=[20,30]
res=conn.execute(query,data).fetchall()
res
```

```python
for rate, books_count in res:
    print(f'{books_count} of books have rate {rate}')
```

```python
conn.close()
```

### Write a sql query to calculate how many books each category has having count > 10.


```python
conn=sqlite3.connect('books.db')
```

```python
query='SELECT category_id,COUNT(id) AS numer_of_books FROM books  GROUP BY category_id HAVING COUNT(id)>? '
data=[10]
res=conn.execute(query,data).fetchall()
res
```

### Write a sql query to get all books with category_name='Music' using subquery.
